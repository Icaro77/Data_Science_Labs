First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.


We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# --- Función para calcular distancia (Haversine) ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# --- Coordenadas de ejemplo ---
launch_site_coord = [28.56367, -80.57163]   # CCAFS SLC 40
city_coord = [28.53834, -81.37924]          # Orlando, FL (ejemplo de ciudad cercana)

# --- Calcular distancia ---
distance = calculate_distance(
    launch_site_coord[0], launch_site_coord[1],
    city_coord[0], city_coord[1]
)

# --- Crear marcador en la ciudad con la distancia ---
distance_marker = folium.Marker(
    location=city_coord,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance)
    )
)

# --- Añadir al mapa ---
site_map.add_child(distance_marker)

# --- Dibujar línea entre el sitio de lanzamiento y la ciudad ---
line = folium.PolyLine(
    locations=[launch_site_coord, city_coord],
    weight=2,
    color='blue'
)
site_map.add_child(line)

# Mostrar el mapa
site_map
```

```python
¿Están los sitios de lanzamiento cerca de ferrocarriles?
Sí, varios de los sitios de SpaceX en Florida (como CCAFS y KSC) están en proximidad a líneas ferroviarias.

Esto tiene sentido logístico: los ferrocarriles permiten transportar cohetes y componentes pesados directamente hasta las instalaciones.

 ¿Están cerca de carreteras/autopistas?
También sí. Los sitios de lanzamiento suelen estar conectados con autopistas principales para facilitar el acceso de personal, equipos y suministros.

En el mapa se observa que las carreteras pasan muy cerca de las plataformas de lanzamiento.

 ¿Están cerca de la costa?
Absolutamente. Todos los sitios de lanzamiento de SpaceX en Florida están a escasos kilómetros del océano Atlántico.

Esto no es casualidad: lanzar hacia el mar reduce riesgos para poblaciones en caso de fallos y permite trayectorias seguras hacia órbitas ecuatoriales.
 ¿Mantienen cierta distancia de las ciudades?
Sí. Aunque están relativamente cerca de ciudades como Orlando o Cocoa Beach, mantienen una distancia prudente.

Esto asegura que las operaciones de lanzamiento no afecten directamente a zonas urbanas densas, pero siguen estando lo suficientemente cerca para aprovechar infraestructura y mano de obra.

 Conclusión
Los sitios de lanzamiento están estratégicamente ubicados:

Muy cerca de costas (seguridad y trayectorias orbitales).

Conectados a ferrocarriles y autopistas (logística).

A cierta distancia de ciudades (seguridad poblacional, pero con acceso a infraestructura).

 En otras palabras, la geografía de los sitios de lanzamiento refleja un equilibrio entre seguridad, logística y eficiencia orbital.
```
